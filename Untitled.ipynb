```python
# Filter rows in the 'Company' column that contain "health" (case-insensitive)
health_related_df = df[df['Company'].str.contains('health', case=False, na=False)]

# Display the filtered DataFrame
print(health_related_df)
```

```python
# Assuming 'Email' is a column in health_related_df
# Remove duplicate entries based on the 'Email' column
unique_health_df = health_related_df.drop_duplicates(subset='Email')

# Display the resulting DataFrame with unique emails
print(unique_health_df)
```

```python
# Export unique_health_df to a CSV file
unique_health_df.to_csv('unique_health_df.csv', index=False)
```
